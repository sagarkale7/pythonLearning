#Numpy
NumPy is short from of numerical Python. Numpy is open source library.

NumPy operations are implemented in C and Fortan, making them faster than equivalent operations perfomed with standard Python lists(objects) especially for the large number of data.

NumPy used as foundation for many other important libraries like Pandas, SciPy, Matplotib, Scikit-learn.


##ndarray object
ndarray object is main feature of numpy whci provide support for multi dimensional array and matrices.

```python
import numpy as np
ndarray_object = np.array([1,2,3])
print(type(ndarray_object))
```

##Arithmetic operations




###Addition

```python
matrix_1 = np.array([1,2,3])
matrix_2 = np.array([4,5,6])

print('Addition of Matrix\n', np.add(matrix_1,matrix_2))
```

###Subtraction





```python
print('\nSubtraction of Matrix\n', np.subtract(matrix_1,matrix_2))
```

###Multiplication

```python
print('\nMultiplication of Matrix\n', np.multiply(matrix_1,matrix_2))
```

###Division

```python
print('Division of Matrix\n', np.divide(matrix_1,matrix_2))
```

###Modulus

```python
print("Modulus (function)\n", np.mod(matrix_1,matrix_2))
```

###Exponentiation

```python
print(np.power(matrix_1,2))
```

##Accessing and modifying NumPy arrays

###Random number array

```python
import numpy as np
random_array = np.random.rand(8)
print("Random number array:\n",random_array)

random_number_matrix = np.random.rand(5,5)

print('\nRandom number array :\n',random_number_matrix)

print('\nFilter based on number of row and column\n', random_number_matrix[0:2,1:3])

random_number_matrix[2:5][ 1:5]= 5

print('\nRandom number array after set:\n',random_number_matrix)
```

###Arange array

```python
arange_array = np.arange(100, 220, 50)
print("Arange array:\n",arange_array)
print("Array filter with condition\n",arange_array[arange_array<=150])
```
